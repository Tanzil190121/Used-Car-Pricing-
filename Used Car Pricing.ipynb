```python
import pandas as pd
```

```python
df=pd.read_csv('autos.csv')
```

```python
df
```

```python
df=df.dropna()
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
first=LabelEncoder()
```

```python
df['name']=first.fit_transform(df['name'])
df['vehicleType']=first.fit_transform(df['vehicleType'])
df['gearbox']=first.fit_transform(df['gearbox'])
df['model']=first.fit_transform(df['model'])
df['fuelType']=first.fit_transform(df['fuelType'])
df['brand']=first.fit_transform(df['brand'])
df['abtest']=first.fit_transform(df['abtest'])
df['yearOfRegistration']=first.fit_transform(df['yearOfRegistration'])
df['notRepairedDamage']=first.fit_transform(df['notRepairedDamage'])
```

```python
df
```

```python
x=df.iloc[:,5:-4]
y=df.iloc[:,5]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.predict(x_test)
```

```python
reg.score(x_test,y_test)
```
